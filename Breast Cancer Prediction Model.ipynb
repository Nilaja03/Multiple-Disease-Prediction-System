Importing the Dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split #splits into training data and testing data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and processing

```python
#loading the data from sklearn 
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
#loading the data to a data frame (panda)
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
#print first five rows of the data frame
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.tail()
```

```python
data_frame.info()
```

```python
data_frame.describe()
```

```python
#checking the distribution of the target variable
data_frame['label'].value_counts()
```

1--->benign 0--->malignant

```python
data_frame.groupby('label').mean()
```

Seperating the features and target

```python
X = data_frame.drop(columns = 'label', axis=1)
Y = data_frame['label']
```

Splitting the data into training data and testing data

```python
#making X and Y as numpy arrays before fitting it to classifier
X = X.values
Y = Y.values
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

model training Logistic Regression

```python
model = LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
#training the logistic regression model using training data
model.fit(X_train, Y_train)
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

Building a predictive System

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print('The Breast cancer is Malignant')

else:
    print('The Breast Cancer is Benign')
```
